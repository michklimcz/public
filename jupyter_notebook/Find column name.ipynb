```python
df = pd.read_excel(r'C:\Users\Micha≈Ç\PycharmProjects\python_public\get_colum_names.xls')
df
```

```python
column_names = []
df['value_to_find'] = df['value_to_find'].replace('ND',-1) #because i want to have only numbers
for id_number in df.id:
    if df[df['id']==id_number].value_to_find.item()>0:
        data_row = df[df['id']==id_number]
        value_to_find = df[df['id']==id_number].value_to_find.item()
        column_names_temp = []
        for column in data_row:
            if data_row[column].item()==value_to_find and column!='value_to_find':
                column_names_temp.append(column)
        column_names.append(column_names_temp)
    else:
        data_row = df[df['id']==id_number]
        value_to_find = df[df['id']==id_number].value_to_find.item()
        column_names.append(str(value_to_find))
       
df['found_column'] = column_names
df
            
                
                
  
            
```
